<span style="font-family:Times New Roman; font-size:1.5em;">Matplotlib.pyplot is a library of functions used to simulate MATLAB software. The pyplot functions enable editabiltity of figures for users operating with the data visualisations of the matplotlib library. The user can use pyplot to format the style of their graphs and utilises plotting with keyword strings if needed. Graph lines can be set to an extensive list of attributes and categorical variables can be used to create plots. By categorical variables here, we mean variables that are selected from a set of pre-defined variable values.

<span style="font-family:Times New Roman; font-size:1.5em;">The following code provides an example of operating with continuous data using a simple pyplot graph. The main defining feature of continuous data is that it can take any value. There is no categorical adherence required when dealing with such data and is perhaps the more common type of data that a data scientist would deal with when measuring values of some physical aspect or activity of an object(s).

```python
import matplotlib.pyplot as plt

plt.plot([-1, -4.5, 16, 23, 78, 22, 3])
plt.show()
```

<span style="font-family:Times New Roman; font-size:1.5em;">The following code is an example of operating with categorical data in pyplot. The data set we are using here is, perhaps the most famous data set used by data analysts, the Iris data set, which is built-in to sklearn, a science data library. The iris data set measures the difference is petal and sepal dimensions of three different families of flower. 

```python
import matplotlib.pyplot as plt

import mpl_toolkits.mplot3d 

from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data[:, :2]  
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)          #Defines limit
plt.ylim(y_min, y_max)          #Defines limit
plt.xticks(())                  #Ignores zero values
plt.yticks(())                  #Ignores zero values

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("Iris Data Set Visualization")
ax.set_xlabel("Sepal Length")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("Sepal Width")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("Petal Length")
ax.w_zaxis.set_ticklabels([])
```

<span style="font-family:Times New Roman; font-size:1.5em;">Continuous data is data that can have any value with a given range for a particular variable relative to some benchmark. Categorical data is data that is fundamentally binary in nature in that it is either an object with a category or not, i.e, it does not have a spectrum of values. These are the main differences between the two data types.
