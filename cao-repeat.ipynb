<span style="font-family:Times New Roman; font-size:1.5em;">The data contained for 2021 on the CAO website was initially a rough CSV with a large amount of null or NaN values for points revision post initial course offers. The data I screened to include highest value over 3 points revisions (where available). Data was formatted by myself in Excel and export as CSV again to import into pandas below. Results given below include CAO points average, highest points and lowest points for a course.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# 2021 CAO Points
df = pd.read_csv('Copy of CAO 2021 - CAO.csv', index_col = 'Subject') #Import CSV to dataframe
df2 = df["Points"].mean()                                             #Calculate average of all points
df3 = df[df["Points"] != 0].max(numeric_only=None)                    #Calculate highest points course
df4 = df[df["Points"] != 0].min(numeric_only=None)                    #Calculate lowest points course
df5 = df["Level"].value_counts()                                      #Count instances of course levels

# 2020 CAO Points
df1 = pd.read_csv('Copy of CAOPointsCharts2020 - CAOPointsCharts2020.csv', index_col = 'Subject') #Import CSV to dataframe
df21 = df1["Points"].mean()                                                                #Calculate average of all points  
df31 = df1[df1["Points"] != 0].max(numeric_only=None)                                       #Calculate highest points course
df41 = df1[df1["Points"] != 0].min(numeric_only=None)                                       #Calculate lowest points course
df51 = df1["Level"].value_counts()                                                         #Count instances of course levels


print(df1)                                                    #Print dataframe                                     
print(""f'Average points over all courses 2020: \n{df21} \n \n' 
        f'Highest points for course 2020: \n{df31} \n \n'
        f'Lowest points for course 2020: \n{df41} \n \n' 
        f'Number of Level 6, 7 and 8 courses respectively 2020: \n{df51[:3]} \n \n'"")

print(df)                                                             #Print dataframe
print(""f'Average points over all courses 2021: \n{df2} \n \n'             
        f'Highest points for course 2021: \n{df3} \n \n'
        f'Lowest points for course 2021: \n{df4} \n \n' 
        f'Number of Level 6, 7 and 8 courses respectively 2021: \n{df5[:3]} \n \n'"")

df['Points'].plot()
plt.xticks([])
plt.title(f"Plot of CAO Points for 2021 per course (Fig.1)")
plt.show()

df1['Points'].plot()
plt.xticks([])
plt.title(f"Plot of CAO Points for 2020 per course (Fig.2)")
plt.show()


# Comparison of CAO averages between 2020 and 2021
a = [df2]
b = [df21]
  
n=1
r = np.arange(n)
width = 0.25
plt.bar(r, a, color = 'b',
        width = width, edgecolor = 'black',
        label='2021')

plt.bar(r + width, b, color = 'g',
        width = width, edgecolor = 'black',
        label='2020')
  
plt.xlabel("Year")
plt.ylabel("Average points for year")
plt.title(f"Average CAO points for 2020 and 2021 (Fig.3)")
plt.legend()
  
plt.show()
```

<span style="font-family:Times New Roman; font-size:1.5em;">We can see from our plots of points by subject for each year (Fig.1 and Fig.2), that there has been an overall increase in 2021 on 2020 in points per course. Even in the outliers of both sets of results, the 2021 outliers are more numerous and tend to even greater extremes at the high-points end. The bulk of the 2020 grades lies in the 150 - 600 points interval of the plot whereas for 2021, the bulk (the majority of courses) lie in the 200 - 600 interval, meaning that the low end has risen, meaning, at first glance, that course points are getting higher overall.

<span style="font-family:Times New Roman; font-size:1.5em;">Though the extremes have certainly risen, the average result for both is almost identical as seen in Fig.3. We can extrapolate from this that the CAO governong body that determines the points for each course, needs those values to fall with a bell-curve value system, that is, that roughly 80% of all course should fall between roughly 200 - 400, and then the percantage of courses lessens at the high and low end-points of the curve, thus allowing the points system to be a closed and dynamic system. By dynamic, we mean that, for example, should a particular course be in high demand, then the screening for that course in terms of points becomes more stringent as the college wants to bring in the person with the highest points relative to the course requirements. This creates a comparative disparity between this and other, less in-demand courses such that the average overall CAO points, i.e, the average chance of getting into a course, remains the same overall.
